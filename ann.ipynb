```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as mlt
```

```python
df = pd.read_pickle('df_train_api.pk')
```

```python
df.head()
```

```python
df0 = df.query('label==0').sample(frac = 0.05)
df = df[df.label != 0]
df = pd.concat([df, df0], ignore_index=True)
X = df[['groups']]
```

```python
y = pd.get_dummies(df['label']).values
```

```python
from  keras.preprocessing.text  import Tokenizer
max_features = 4000
tokenizer = Tokenizer(num_words=max_features, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', split=' ', lower=True, )
tokenizer.fit_on_texts(X['groups'].values)

X = tokenizer.texts_to_sequences(X['groups'].values)

# add padding
from keras.preprocessing.sequence import pad_sequences
X = pad_sequences(X, maxlen=40)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
clf = Sequential()
clf.add(Dense(output_dim = 20, kernel_initializer='uniform', activation='relu', input_dim = 40))
clf.add(Dropout(rate=0.1))
clf.add(Dense(output_dim = 20, activation='relu', kernel_initializer='uniform'))
clf.add(Dropout(rate=0.3))
clf.add(Dense(6,kernel_initializer='uniform', activation='softmax'))
```

```python
clf.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
clf.fit(X_train, y_train, batch_size=10, epochs=500)
```

```python
from keras.models import model_from_json
model_json = clf.to_json()
with open("firstann.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
clf.save_weights("model.h5")
```

```python
def model(dropout_rate=0.0):
    clf = Sequential()
    clf.add(Dense(output_dim = 20, kernel_initializer='uniform', activation='relu', input_dim = 40))
    clf.add(Dropout(rate=0.1))
    clf.add(Dense(output_dim = 20, activation='relu', kernel_initializer='uniform'))
    clf.add(Dropout(rate=0.3))
    clf.add(Dense(6,kernel_initializer='uniform', activation='softmax'))
    clf.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])
    return clf


    
```

```python

from keras.wrappers.scikit_learn import KerasClassifier
model = KerasClassifier(build_fn=model, dropout_rate=0.2)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = dict(epochs=[100,200,300,400,500])
grid = GridSearchCV(estimator=model, param_grid=param_grid)
grid_result = grid.fit(X_train, y_train)
```
